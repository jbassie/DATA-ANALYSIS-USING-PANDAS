```python
top_10_countries = data.Team.value_counts().sort_values(ascending = False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title('Overall Participation by country')
#plt.xticks(rotation =20)
sns.barplot(x=top_10_countries.index, y = top_10_countries, palette = 'Set2')
```

```python
plt.figure(figsize=(12,6))
plt.title('Age Distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(data.Age, bins =np.arange(10,80,2), color = 'Orange', edgecolor = 'white');
```

```python
winter_sports = data[data.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
summer_sports = data[data.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
gender_counts = data.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle = 180, shadow  = True)
```

```python
Medals = data.Medal.value_counts()
Medals
```

```python
plt.figure(figsize = (15,10))
plt.title('Medal Distribution')
plt.pie(Medals , labels = Medals.index, autopct = '%1.1f%%', shadow = True, startangle = 180)
```

```python
womenOlympics = data[(data.Sex == 'F') & (data.Season =='Summer')]
womenOlympics
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize = (20,10))
sns.countplot(x='Year', data= womenOlympics, palette = 'Spectral')
plt.title('Women Participation')
```
