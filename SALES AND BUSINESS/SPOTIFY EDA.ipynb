##### Show the correlated Heatmap

```python
corr_df = df_track.corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot  =True, fmt ='.1g', vmin=-1, vmax = 1, center =0, cmap= 'inferno', linewidths= 1, linecolor = 'Black')
heatmap.set_title("Correlation HeatMap Between Varible")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

##### REGRESSOIN PLOT

```python
sample_df = df_track.sample(int(0.004* len(df_track)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title = "Loudiness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'b').set(title = "Popularity vs Accousticness Correlation")
```

```python
#Duration of songs according to different music genre
plt.figure(figsize = (15,8))
plt.title('Duration of songs in Different Genre')
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y='genre', x= 'duration_ms', data= df_track)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (15,8))
famous = df_track.sort_values('popularity', ascending = False).head(10)
sns.barplot(y='genre', x = 'popularity', data = famous).set(title = 'Top 5 Genres by popularity')
```
