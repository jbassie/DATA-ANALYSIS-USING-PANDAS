```python
womenOlympics =data[(data.Sex== 'F') & (data.Season =='Summer')]
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize = (20,10))
sns.countplot(x = "Year", data = womenOlympics, palette = "Spectral")
plt.title('Women Particiaption')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize= (20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
gold_medals = data[(data.Medal =='Gold')]
gold_medals.head()
```

```python
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
gold_medals['ID'][gold_medals['Age'] > 60].count()
```

```python
sporting_events= gold_medals['Sport'][gold_medals['Age']>60]
sporting_events
```

```python
plt.figure(figsize = (10,5))
plt.tight_layout()
sns.countplot(sporting_events)
plt.title("Gold Medals for athlete over 60 years")
```

```python
gold_medals.Region.value_counts().reset_index(name = 'Medal').head()
```

```python
plt.figure(figsize = (20,10))
totalGoldMedals = gold_medals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x='index', y='Medal', data = totalGoldMedals, height = 5, kind = 'bar', palette = 'rocket')
g.despine(left = True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels('Number of Medals')
plt.title('Gold MEdal per Country')
```
